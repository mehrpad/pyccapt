```python
import re

def parse_events(file_path):
    events = []
    current_event = None
    event_started = False
    
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            
            # Check if this line starts a new event block
            event_header_match = re.match(r'------- #(\d+) -------', line)
            if event_header_match:
                # If there was a previous event, store it
                if current_event is not None:
                    events.append(current_event)
                
                # Start a new event
                event_number = int(event_header_match.group(1))
                current_event = {'event_number': event_number, 'channels': []}
                event_started = True
                continue
            
            # If we are within an event block
            if event_started and current_event is not None:
                # Extract T and dT values
                if line.startswith('T'):
                    t_match = re.search(r'T\s+=\s+([\d.]+)\s+ns', line)
                    if t_match:
                        current_event['T'] = float(t_match.group(1))
                
                elif line.startswith('dT'):
                    dt_match = re.search(r'dT\s+=\s+([\d.]+)\s+ns', line)
                    if dt_match:
                        current_event['dT'] = float(dt_match.group(1))
                
                # Extract channel information
                channel_match = re.match(r'chan\s+(\d+)\s+(\d+)\s+(.*)', line)
                if channel_match:
                    channel_num = int(channel_match.group(1))
                    num_values = int(channel_match.group(2))
                    values_str = channel_match.group(3).strip()
                    
                    # Split the values string based on the number of values
                    if num_values == 1:
                        values = [float(values_str)]
                    elif num_values >= 2:
                        values = [float(val) for val in values_str.split()]
                    else:
                        values = []
                    
                    current_event['channels'].append({
                        'channel': channel_num,
                        'num_values': num_values,
                        'values': values
                    })
    
    # Append the last event if there was one
    if current_event is not None:
        events.append(current_event)
    
    return events

# Example usage:
events = parse_events(raw_filename)
```

```python
# Initialize dictionaries to count occurrences
def initialize_counters():
    return {
        'total': {i: 0 for i in range(1, 500)},
        'invalid': {i: 0 for i in range(1, 500)},
        'dld2': {i: 0 for i in range(1, 500)},
        'dld4': {i: 0 for i in range(1, 500)},
        'dld6': {i: 0 for i in range(1, 500)}
    }

def check_dld_patterns(ch_counts):
    """Check for different DLD patterns and return the highest valid pattern found"""
    ch1, ch2, ch3, ch4, ch5, ch6 = [ch_counts.get(i, 0) for i in range(1, 7)]
    
    # Check for 6DLD pattern
    if ch1 >= 1 and ch2 >= 1 and ch3 >= 1 and ch4 >= 1 and ch5 >= 1 and ch6 >= 1:
        return 6
    
    # Check for 4DLD patterns
    if ((ch1 >= 1 and ch2 >= 1 and ch3 >= 1 and ch4 >= 1) or 
        (ch1 >= 1 and ch2 >= 1 and ch5 >= 1 and ch6 >= 1) or 
        (ch3 >= 1 and ch4 >= 1 and ch5 >= 1 and ch6 >= 1)):
        return 4
    
    # Check for 2DLD patterns
    if ((ch1 >= 1 and ch2 >= 1) or 
        (ch3 >= 1 and ch4 >= 1) or 
        (ch5 >= 1 and ch6 >= 1)):
        return 2
    
    return 0  # No valid pattern found

def extract_max_dld_patterns(ch_counts, total_values):
    """Extract the maximum number of DLD patterns of each type"""
    patterns = {'dld2': 0, 'dld4': 0, 'dld6': 0}
    
    # Make a copy of the counts to manipulate
    counts = ch_counts.copy()
    
    # First try to extract as many 6DLD patterns as possible
    while True:
        if all(counts.get(i, 0) >= 1 for i in range(1, 7)):
            patterns['dld6'] += 1
            for i in range(1, 7):
                counts[i] -= 1
        else:
            break
    
    # Then try to extract 4DLD patterns
    while True:
        if ((counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1 and 
             counts.get(3, 0) >= 1 and counts.get(4, 0) >= 1) or
            (counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1 and 
             counts.get(5, 0) >= 1 and counts.get(6, 0) >= 1) or
            (counts.get(3, 0) >= 1 and counts.get(4, 0) >= 1 and 
             counts.get(5, 0) >= 1 and counts.get(6, 0) >= 1)):
            
            patterns['dld4'] += 1
            
            # Choose which pattern to use based on availability
            if counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1 and counts.get(3, 0) >= 1 and counts.get(4, 0) >= 1:
                counts[1] -= 1
                counts[2] -= 1
                counts[3] -= 1
                counts[4] -= 1
            elif counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1 and counts.get(5, 0) >= 1 and counts.get(6, 0) >= 1:
                counts[1] -= 1
                counts[2] -= 1
                counts[5] -= 1
                counts[6] -= 1
            else:
                counts[3] -= 1
                counts[4] -= 1
                counts[5] -= 1
                counts[6] -= 1
        else:
            break
    
    # Finally, extract 2DLD patterns
    while True:
        if ((counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1) or
            (counts.get(3, 0) >= 1 and counts.get(4, 0) >= 1) or
            (counts.get(5, 0) >= 1 and counts.get(6, 0) >= 1)):
            
            patterns['dld2'] += 1
            
            # Choose which pattern to use
            if counts.get(1, 0) >= 1 and counts.get(2, 0) >= 1:
                counts[1] -= 1
                counts[2] -= 1
            elif counts.get(3, 0) >= 1 and counts.get(4, 0) >= 1:
                counts[3] -= 1
                counts[4] -= 1
            else:
                counts[5] -= 1
                counts[6] -= 1
        else:
            break
    
    return patterns

def length_count(events):
    counters = initialize_counters()
    
    for event in events:
        # Create a dictionary of channel counts
        ch_counts = {}
        for ch in event['channels']:
            ch_counts[ch['channel']] = ch['num_values']
        
        # Calculate total values
        total_values = sum(ch_counts.get(i, 0) for i in range(1, 7))
        
        if total_values == 0:
            continue
        
        # Count this event in the total
        counters['total'][total_values] += 1
        
        # Extract patterns for this event
        patterns = extract_max_dld_patterns(ch_counts, total_values)
        
        # Record the patterns found
        if sum(patterns.values()) > 0:
            for pattern_type, count in patterns.items():
                if count > 0:
                    if pattern_type == 'dld2':
                        counters['dld2'][total_values] += count
                    elif pattern_type == 'dld4':
                        counters['dld4'][total_values] += count
                    elif pattern_type == 'dld6':
                        counters['dld6'][total_values] += count
        else:
            # If no patterns were found, mark as invalid
            counters['invalid'][total_values] += 1
        
        # Store the detected DLD patterns in the event
        event['dlts'] = []
        for _ in range(patterns['dld6']):
            event['dlts'].append(6)
        for _ in range(patterns['dld4']):
            event['dlts'].append(4)
        for _ in range(patterns['dld2']):
            event['dlts'].append(2)
    
    return events, counters

def plot_results(counters, max_bins=20):
    # Plotting
    bins = np.arange(1, max_bins + 1)
    
    # Convert dictionaries to arrays for plotting
    total_arr = np.array([counters['total'].get(i, 0) for i in bins])
    dld2_arr = np.array([counters['dld2'].get(i, 0) for i in bins])
    dld4_arr = np.array([counters['dld4'].get(i, 0) for i in bins])
    dld6_arr = np.array([counters['dld6'].get(i, 0) for i in bins])
    
    # Define the width of the bars
    width = 0.2
    fig, ax = plt.subplots(figsize=(7, 2.5))
    
    # Plot bar charts for all arrays
    plt.bar(bins, total_arr, width=width*3, label='Frequency', alpha=0.5, color='gray')
    plt.bar(bins - width, dld2_arr, width=width, label='2DLTS', color='orange')
    plt.bar(bins, dld4_arr, width=width, label='4 DLTS', color='blue')
    plt.bar(bins + width, dld6_arr, width=width, label='6 DLTS', color='green')
    
    ax.set_xlabel('Number of Delayline Time Stamps per Pulse', fontsize=10)
    plt.xticks(bins, [str(i) for i in bins])
    ax.set_ylabel('Count', fontsize=10)
    ax.set_yscale('log')
    plt.legend(loc='upper right')
    plt.tight_layout()
    
    # Set font type for SVG
    from matplotlib import rcParams
    rcParams['svg.fonttype'] = 'none'
    
    plt.savefig("raw_stat_roetdek.svg", format="svg", dpi=600)
    plt.savefig("raw_stat_roetdek.png", format="png", dpi=600)
    plt.show()


events_dlts, counters = length_count(events)
plot_results(counters)
```
